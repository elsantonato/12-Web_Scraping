```python
# import dependencies
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import pymongo
import matplotlib.pyplot as plt
from splinter import Browser

# Output File (CSV)
output_data_file = "csv_outputs/home_depot_carpenter.csv"
```

```python
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
```

```python
browser = Browser("chrome", **executable_path, headless=False)
```

```python
#set the url as a variable
url = 'https://www.homedepot.ca/en/home/search.html?q=carpenter#!q=carpenter'
browser.visit(url)
```

```python
#parse the text version of the response using an html parser 
html = browser.html
soup = bs(html, 'html.parser')
```

```python
#pretty print the parsed text
print(soup.prettify())
```

```python
results = soup.find_all('section', class_="hdca-product-box__main")
for item in results:
    print(item)
```

```python
results = soup.find_all('section', class_="hdca-product-box__main")
title_list = []
price_list = []
#link_list = []

#loop through returned results
for result in results:
    #error handling
    try: 
        #identify and return title of listing
        title = result.find('span', itemprop="name").text.strip()
        #identify and return price of listing
        price = result.find('span', itemprop="price").text.strip()
        #identify and return link to listing
        #link = 'https://www.homedepot.ca' + result.find('a', itemprop="url")['href']
        #link = f'https://www.homedepot.ca'/{result.find('a', itemprop="url")['href']}
        
        title_list.append(title)
        price_list.append(price)
        #link_list.append(link)
        
    except AttributeError as e:
        print(e)
```

```python
#df = pd.DataFrame({"title": title_list, "price": price_list, "link": link_list})
df = pd.DataFrame({"title": title_list, "price": price_list})
```
