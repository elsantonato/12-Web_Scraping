```python
# import dependencies
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import pymongo
import matplotlib.pyplot as plt
```

```python
#set the url as a variable
url = 'https://www.mycarpentry.com/carpentry-tools.html'
```

```python
#make a call to the internet page
response = requests.get(url)
```

```python
#parse the text version of the response using an html parser 
soup = bs(response.text, 'html.parser')

#pretty print the parsed text
print(soup.prettify())
```

```python
results = soup.find_all('div', class_="info")
for item in results:
    print(item)
```

```python
results = soup.find_all('div', class_="info")
#loop through returned results
for result in results:
    #error handling
    try: 
        #identify and return title of listing
        title = result.find('div', class_= "info").text
        #identify and return price of listing
        price = result.a.span.text
        #identify and return link to listing
        link = result.a['href']
        
        #print results only if title, price, and link are available
        if (title and price and link):
            print('------------')
            print(title)
            print(price)
            print(link)
    except AttributeError as e:
        print(e)
```
