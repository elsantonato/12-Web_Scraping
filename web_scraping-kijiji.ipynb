```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

# Output File (CSV)
output_data_file = "csv_outputs/kijiji_carpenter_tools.csv"
```

```python
url = 'https://www.kijiji.ca/b-city-of-toronto/carpenter-tools/k0l1700273?dc=true'
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('div', class_="info-container")
for item in results:
    print(item)
```

```python
results = soup.find_all('div', class_="info-container")
title_list = []
price_list = []
link_list = []

#loop through returned results
for result in results:
    #error handling
    try: 
        #identify and return title of listing
        title = result.find('a', class_='title').text.strip()
        #identify and return price of listing
        price = result.find('div', class_="price").text.strip()
        #identify and return link to listing
        link = 'https://www.kijiji.ca' + result.find('a', class_='title')['href']
        
        title_list.append(title)
        price_list.append(price)
        link_list.append(link)
        
    except AttributeError as e:
        print(e)
```

```python
df = pd.DataFrame({"title": title_list, "price": price_list, "link": link_list})
```
